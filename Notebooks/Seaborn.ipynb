#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
tips
```

```python
sns.lineplot(x='size',y='total_bill' ,data=tips)
plt.title('Line plot of Total bill vs size')
plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.title('Line plot of Total_bill vs tip')
plt.show()
```

```python
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips) #colour diffrentior as hue
plt.title("scatter plot of total_bill vs tip")
plt.show()
```

```python
tips
```

```python
tips.head()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("bar graph of total_bill vs day")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box plot")
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'] ,bins=5,kde=True) #kde=true/false-->true shows normal distribution curve.
```

image.png

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True) #fill is fill the distribution.
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
## HEatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm') # annot=True -->shows data in each cell of heatmap.
#cmap-->color of map.
#masking is done using numpy.
```
