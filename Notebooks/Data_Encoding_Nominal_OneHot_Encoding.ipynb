### Nominal/OHE Encoding
One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

1. Red: [1, 0, 0]
2. Green: [0, 1, 0]
3. Blue: [0, 0, 1]

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe 
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
##create an instance of Onehotencoder
encoder=OneHotEncoder()
```

```python
## perform fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python

encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
## for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
import seaborn as sns
```

```python
df2=sns.load_dataset('tips')
```

```python
df2.head()
```

```python

categorical_data = ['sex', 'smoker', 'day', 'time']
encoded_data = {}

for x in categorical_data:
    encoded_data[x] = encoder.fit_transform(df2[[x]]).toarray()  
```

```python
encoded_df2s = []  

for x in categorical_data:
    encoded_array = encoded_data[x]  
    column_names = [f"{x}_{i}" for i in range(encoded_array.shape[1])]  
    encoded_df2s.append(pd.DataFrame(encoded_array, columns=column_names, index=df2.index))  
```

```python
encoded_df2s
```

```python
df2_encoded = pd.concat([df2.drop(columns=categorical_data)] + encoded_df2s, axis=1)
```
